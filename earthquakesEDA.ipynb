## status

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
df['status'].value_counts().plot(kind='bar')
plt.title('Distribution of Earthquake Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

## tsunami

```python
tsunami_counts = df['tsunami'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(tsunami_counts, labels=tsunami_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Tsunami Occurrence in Earthquakes')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='state', hue='status')
plt.title('Earthquake Status by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Status')
plt.show()
```

## heatmap of 'status' vs. 'tsunami'

```python
status_tsunami_crosstab = pd.crosstab(df['status'], df['tsunami'])
sns.heatmap(status_tsunami_crosstab, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Relationship between Status and Tsunami')
plt.xlabel('Tsunami')
plt.ylabel('Status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['depth'], bins=20, edgecolor='k')
plt.title('Depth Distribution of Earthquakes')
plt.xlabel('Depth (km)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df['time'] = pd.to_datetime(df['time'])
df['year'] = df['time'].dt.year
earthquakes_by_year = df.groupby('year').size()
plt.figure(figsize=(12, 6))
earthquakes_by_year.plot(kind='line', marker='o')
plt.title('Earthquakes Over Time (by Year)')
plt.xlabel('Year')
plt.ylabel('Number of Earthquakes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='magnitudo', y='depth', alpha=0.5)
plt.title('Scatter Plot of Magnitude vs. Depth')
plt.xlabel('Magnitude')
plt.ylabel('Depth (km)')
plt.grid(True)
correlation = df['magnitudo'].corr(df['depth'])
print(f'Correlation coefficient between Magnitude and Depth: {correlation:.2f}')

plt.show()
```

## geospecific analysis

```python
import pandas as pd
import folium
map_center = [df['latitude'].mean(), df['longitude'].mean()]
mymap = folium.Map(location=map_center, zoom_start=4)
for index, row in df.iterrows():
    folium.Marker([row['latitude'], row['longitude']], tooltip=row['place']).add_to(mymap)
mymap.save('earthquake_map.html')
```

## correlation analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Eartquakes-1990-2023.csv')
correlation_matrix = df[['magnitudo', 'depth']].corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
